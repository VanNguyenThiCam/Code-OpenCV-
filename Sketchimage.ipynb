```python
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)) #chuyển ảnh về màu thật theo opencv 
```

```python
grey_img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY) #chuyển về ảnh xám 
```

```python

invert = cv2.bitwise_not(grey_img)
```

```python
blur = cv2.GaussianBlur(invert,(21,21),0)
invertedblur = cv2.bitwise_not(blur)
sketch = cv2.divide(grey_img, invertedblur, scale = 256.0)
```

```python
cv2.imwrite('/content/gdrive/MyDrive/Colab Notebooks/Data/sketch.jpg',sketch)
```
